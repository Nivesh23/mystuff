```python
import pandas as pd
```

```python
housing_pd=pd.read_csv("sample_data/DS_test.csv.csv")
```

```python
housing_pd
```

```python
housing_pd_shuffled=housing_pd.sample(n=len(housing_pd),random_state=1)
housing_pd_shuffled
```

```python
housing_pd_final=housing_pd[['Transaction date','House Age','Distance from nearest Metro station (km)','Number of convenience stores','latitude','longitude','Number of bedrooms','House size (sqft)','House price of unit area']]
```

```python
housing_pd_final
```

```python
housing_pd_final=housing_pd_final.dropna()
len(housing_pd_final)
```

```python
train_pd,test_pd,val_pd=housing_pd_final[:300],housing_pd_final[300:400],housing_pd_final[400:]
len(train_pd),len(test_pd),len(val_pd)
```

```python
X_train,y_train=train_pd.to_numpy()[:,:-1],train_pd.to_numpy()[:,-1]
X_val,y_val=val_pd.to_numpy()[:,:-1],val_pd.to_numpy()[:,-1]
X_test,y_test=test_pd.to_numpy()[:,:-1],test_pd.to_numpy()[:,-1]
X_train.shape,y_train.shape,X_val.shape,y_val.shape,X_test.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
scaler=StandardScaler().fit(X_train[:, :9])

```

```python
def preprocessor(X):
  A=np.copy(X)
  A[:, :9]=scaler.transform(A[:, :9])
  return A
X_train=preprocessor(X_train)
X_val=preprocessor(X_val)
X_test=preprocessor(X_test)


```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.linear_model import LinearRegression
lm=LinearRegression().fit(X_train,y_train)
mse(lm.predict(X_train),y_train,squared=False),mse(lm.predict(X_val),y_val,squared=False)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=2).fit(X_train,y_train)
mse(knn.predict(X_train),y_train,squared=False),mse(knn.predict(X_val),y_val,squared=False)
```

```python

from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(max_depth=10).fit(X_train,y_train)
mse(rfr.predict(X_train),y_train,squared=False),mse(rfr.predict(X_val),y_val,squared=False)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gfr=GradientBoostingRegressor(n_estimators=20).fit(X_train,y_train)
mse(gfr.predict(X_train),y_train,squared=False),mse(gfr.predict(X_val),y_val,squared=False)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import RootMeanSquaredError

simple_nm=Sequential()
simple_nm.add(InputLayer((8,)))
simple_nm.add(Dense(2,'relu'))
simple_nm.add(Dense(1,'linear'))

opt=Adam(learning_rate=10)
cp=ModelCheckpoint('models/simple_nm',save_best_only=True)
simple_nm.compile(optimizer=opt,loss='mse',metrics=[RootMeanSquaredError()])
simple_nm.fit(x=X_train,y=y_train,validation_data=(X_val,y_val),callbacks=[cp],epochs=100)
```

```python
from tensorflow.keras.models import load_model

simple_nm=load_model('models/simple_nm')
mse(simple_nm.predict(X_train),y_train,squared=False),mse(simple_nm.predict(X_val),y_val,squared=False)
```

```python

medium_nm=Sequential()
medium_nm.add(InputLayer((8,)))
medium_nm.add(Dense(32,'relu'))
medium_nm.add(Dense(16,'relu'))
medium_nm.add(Dense(1,'linear'))

opt=Adam(learning_rate=10)
cp=ModelCheckpoint('models/medium_nm',save_best_only=True)
medium_nm.compile(optimizer=opt,loss='mse',metrics=[RootMeanSquaredError()])
medium_nm.fit(x=X_train,y=y_train,validation_data=(X_val,y_val),callbacks=[cp],epochs=100)
```

```python
medium_nm=load_model('models/medium_nm')
mse(medium_nm.predict(X_train),y_train,squared=False),mse(medium_nm.predict(X_val),y_val,squared=False)
```

```python
large_nm=Sequential()
large_nm.add(InputLayer((8,)))

large_nm.add(Dense(64,'relu'))
large_nm.add(Dense(32,'relu'))
large_nm.add(Dense(1,'linear'))

opt=Adam(learning_rate=10)
cp=ModelCheckpoint('models/large_nm',save_best_only=True)
large_nm.compile(optimizer=opt,loss='mse',metrics=[RootMeanSquaredError()])
large_nm.fit(x=X_train,y=y_train,validation_data=(X_val,y_val),callbacks=[cp],epochs=100)
```

```python
large_nm=load_model('models/large_nm')
mse(large_nm.predict(X_train),y_train,squared=False),mse(medium_nm.predict(X_val),y_val,squared=False)
```
